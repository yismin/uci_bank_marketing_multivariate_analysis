```python
# Numerical variables
num_cols = df.select_dtypes(include='number')
print(num_cols.columns)

# Categorical variables
cat_cols = df.select_dtypes(include='object')
print(cat_cols.columns)
```

```python
import warnings
warnings.filterwarnings('ignore')

num_cols_list = num_cols.columns.tolist()
n = len(num_cols_list)
rows = math.ceil(n / 4)

plt.figure(figsize=(4*4, rows*3))
for i, col in enumerate(num_cols_list, 1):
    plt.subplot(rows, 4, i)
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.tight_layout()

plt.show()
```

```python
cat_cols_list = cat_cols.columns.tolist()  # replace with your categorical columns variable
n = len(cat_cols_list)
rows = math.ceil(n / 4)
plt.figure(figsize=(4*4, rows*3))
for i, col in enumerate(cat_cols_list, 1):
    plt.subplot(rows, 4, i)
    sns.countplot(data=df,x=col)
    plt.title(col, fontsize=10)
    plt.xticks(rotation=30, ha='right')
    plt.tight_layout()

plt.show()
```
